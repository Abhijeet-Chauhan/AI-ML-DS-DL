```python
##model training
from sklearn.model_selection import train_test_split

df = pd.read_csv('Algerian_forest_fires_dataset_cleaned.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day','month','year'],axis=1, inplace=True)
```

```python
##Encoding of categories
df['Classes']=np.where(df["Classes"].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
#Train and Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
## Feature Selection
X_train.corr()
```

```python
## Check for multicollinearity
plt.Figure(figsize=(10, 8))
corr= X_train.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]>threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.85)
```

```python
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
```

```python
##feture scaling or standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_trsin_scaled = scaler.fit_transform(X_train)
```

```python
# Ensure X_test has the same columns as X_train before scaling
X_test_aligned = X_test[X_train.columns]
X_test_scaled = scaler.transform(X_test_aligned)
```

```python
plt.subplots(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('Distribution of Features in Training Set')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_trsin_scaled)
plt.title('Distribution of Scaled Features in Training Set')
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
model = LinearRegression()
```

```python
model.fit(X_trsin_scaled, y_train)
y_pred = model.predict(X_test_scaled )
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
```

```python
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {score}')
```

```python
plt.scatter(y_test, y_pred)
```

```python
## rideg and lasso regression
from sklearn.linear_model import Ridge, Lasso
Lasso_model = Lasso()
Lasso_model.fit(X_trsin_scaled, y_train)
y_pred = Lasso_model.predict(X_test_scaled )
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {score}')
```

```python
## rideg and lasso regression
from sklearn.linear_model import Ridge, Lasso
ridege_model = Ridge()
ridege_model.fit(X_trsin_scaled, y_train)
y_pred = ridege_model.predict(X_test_scaled )
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {score}')
```

```python
##elasticnet
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
elastic.fit(X_trsin_scaled, y_train)
y_pred = elastic.predict(X_test_scaled )
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {score}')
```
